```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# 1. Numerical Data Plotting
- relplot()
- scatterplot()
- lineplot()

```python
sns.set(style="darkgrid")
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.relplot(x="total_bill", y="tip", data=tips)
```

```python
tips["smoker"].value_counts()
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, hue="smoker", style="time")
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, hue="smoker", style="size")
```

```python
df = pd.DataFrame(dict({"time": np.arange(500), "value": np.random.randn(500).cumsum()}))
df.head()
```

```python
sns.relplot(x="time", y="value", data=df, kind="line", sort=True)
```

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
sns.relplot(x="timepoint", y="signal", data=fmri, kind="line")
```

```python
sns.relplot(x="timepoint", y="signal", data=fmri, kind="line", ci=False)
```

```python
sns.relplot(x="timepoint", y="signal", hue="event", data=fmri,  kind="line")
```

```python
sns.relplot(x="timepoint", y="signal", hue="region", style="event", data=fmri, kind="line")
```

```python
sns.relplot(x="timepoint", y="signal", hue="region", style="event", data=fmri, kind="line", marker=True)
```

```python
sns.relplot(x="timepoint", y="signal", hue="region", units="subject", data=fmri.query("event=='stim'"), estimator=None, kind="line")
```

```python
dots = sns.load_dataset("dots")
dots.head()
```

```python
sns.relplot(x="time", y="firing_rate", data=dots, kind="line")
```

```python
palette = sns.cubehelix_palette(light=0.5, n_colors=6)
sns.relplot(x="time", y="firing_rate", hue="coherence", kind="line", size="coherence", sizes=(1, 5), data=dots, ci=False, palette=palette)
```

```python
df = pd.DataFrame(dict(time=pd.date_range("2020-01-01", periods=500), value=np.random.randn(500).cumsum()))
df.head()
```

```python
g = sns.relplot(x="time", y="value", kind="line", data=df)
g.fig.autofmt_xdate()
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", col="time", data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", col="size", data=tips)
```

```python
fmri.head()
```

```python
sns.relplot(x="timepoint", y="signal", hue="subject", kind="line", col="event", row="region", ci=False, data=fmri)
```

```python
sns.scatterplot(x="total_bill", y="tip", hue="smoker", data=tips)
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
sns.relplot(x="sepal_length", y="petal_length", data=iris)
```

```python
sns.lineplot(x="timepoint", y="signal", hue="event", data=fmri)
```

```python
sns.lineplot(x="timepoint", y="signal", hue="event", err_style="bars", data=fmri)
```

# 2. Categorical Data Plotting 
- catplot()
- stripplot()
- swarmplot()
- boxplot()
- boxenplot()
- violinplot()
- barplot()
- pointplot()

```python
tips.head()
```

```python
sns.catplot(x="day", y="total_bill", data=tips)
```

```python
sns.catplot(y="day", x="total_bill", data=tips)
```

```python
sns.catplot(x="day", y="total_bill", data=tips, jitter=False)
```

```python
sns.catplot(x="day", y="total_bill", data=tips, kind="swarm", hue="sex")
```

```python
sns.catplot(x="smoker", y="total_bill", data=tips)
```

```python
sns.catplot(x="smoker", y="total_bill", data=tips, order=["No", "Yes"])
```

```python
sns.catplot(x="day", y="total_bill", data=tips, kind="box", hue="sex")
```

```python
sns.catplot(x="day", y="total_bill", data=tips, kind="box", hue="sex", dodge=False)
```

```python
diamonds = sns.load_dataset("diamonds")
diamonds.head()
```

```python
sns.catplot(x="color", y="price", data=diamonds, kind="boxen")
```

```python
sns.catplot(x="color", y="price", data=diamonds.sort_values(by="color"), kind="boxen")
```

```python
tips.head()
```

```python
sns.catplot(x="day", y="total_bill", data=tips, kind="boxen")
```

```python
sns.catplot(y="day", x="total_bill", data=tips, kind="violin", hue="sex")
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
sns.catplot(x="sex", y="survived", data=titanic, hue="class", kind="bar")
```

```python
sns.catplot(x="deck", y="survived", data=titanic, kind="bar", palette="ch:0.5")
```

```python
sns.catplot(x="deck", y="survived", data=titanic, kind="bar", hue="class", palette="ch:0.5")
```

```python
sns.catplot(x="deck", y="survived", data=titanic, kind="point", hue="class", palette="ch:0.5")
```
