```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# 1. Numerical Data Plotting

```python
sns.set(style="darkgrid")
```

## relplot()

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.relplot(x="total_bill", y="tip", data=tips)
```

```python
tips["smoker"].value_counts()
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, hue="smoker", style="time")
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, hue="smoker", style="size")
```

```python
df = pd.DataFrame(dict({"time": np.arange(500), "value": np.random.randn(500).cumsum()}))
df.head()
```

```python
sns.relplot(x="time", y="value", data=df, kind="line", sort=True)
```

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
sns.relplot(x="timepoint", y="signal", data=fmri, kind="line")
```

```python
sns.relplot(x="timepoint", y="signal", data=fmri, kind="line", ci=False)
```

```python
sns.relplot(x="timepoint", y="signal", hue="event", data=fmri,  kind="line")
```

```python
sns.relplot(x="timepoint", y="signal", hue="region", style="event", data=fmri, kind="line")
```

```python
sns.relplot(x="timepoint", y="signal", hue="region", style="event", data=fmri, kind="line", marker=True)
```

```python
sns.relplot(x="timepoint", y="signal", hue="region", units="subject", data=fmri.query("event=='stim'"), estimator=None, kind="line")
```

```python
dots = sns.load_dataset("dots")
dots.head()
```

```python
sns.relplot(x="time", y="firing_rate", data=dots, kind="line")
```

```python
palette = sns.cubehelix_palette(light=0.5, n_colors=6)
sns.relplot(x="time", y="firing_rate", hue="coherence", kind="line", size="coherence", sizes=(1, 5), data=dots, ci=False, palette=palette)
```

```python
df = pd.DataFrame(dict(time=pd.date_range("2020-01-01", periods=500), value=np.random.randn(500).cumsum()))
df.head()
```

```python
g = sns.relplot(x="time", y="value", kind="line", data=df)
g.fig.autofmt_xdate()
```
